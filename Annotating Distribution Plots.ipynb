```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
url = 'https://docs.google.com/spreadsheets/d/1VMaw2oCn0ABitd-alLAEsEhGS1Je2UFNLu76TKrIH7w/gviz/tq?tqx=out:csv&sheet=Raw_Medical_Data_for_day1'
df = pd.read_csv(url, index_col = 0)
df.head()
```

# Obtain the histogram of Income



```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(df["Income"], stat="probability");
```

```python
## Calculate the Mean
mean = df["Income"].mean()
mean
```

## Add the line for the mean

```python
fig, ax  = plt.subplots(figsize=(10,6))
#think about assigning df["Income"] to a variable for easy switch out (variable = df{col})
sns.histplot(df["Income"], stat="probability")
#Putting the actual mean value in axvline so it can show up in the legend
ax.axvline(mean, color="red", label=f"Mean = {mean:,.2f}");
#could use f"{col}"
ax.set_title("Income")
ax.legend();
```

# Creating Mean and Median Lines

```python
col = "Income"
feature = df[col]
mean = feature.mean()
median = feature.median()
fig, ax = plt.subplots(figsize=(10,6))
sns.histplot(feature, stat="probability")
ax.axvline(mean, color="red", label = f"Mean = {mean:,.2f}")
ax.axvline(median,ls = "--", color="green", label = f"Median = {median:,.2f}")
ax.set_title(f"{col}")
ax.legend();
```

# Highlight a range of values

```python
col = "Income"
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig, ax = plt.subplots(figsize=(10,6))
sns.histplot(feature, stat="probability")
ax.axvline(plus_one_std , color="black", label=f"+1 std {plus_one_std:,.2f}")
ax.axvline(minus_one_std, color="black", label=f"-1 std {minus_one_std:,.2f}")
ax.set_title(f"{col}")
ax.legend();
```

# Highlight Space btw 2 lines

```python
col = 'Income'
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
# to have yellow area behind the rest of the plot use "zorder = 0"
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder=0)
ax.set_title(f'{col}')
ax.legend();
```
